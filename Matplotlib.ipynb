:Matplotlib is a powerful and popular Python library used for creating static, interactive, and publication-quality visualizations. It provides a wide array of functions for generating plots, histograms, bar charts, scatterplots, and more, making it an essential tool for data scientists, researchers, and engineers working with data visualization in Python.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#sample code that generates two arrays of 50 random numbers between 0 and 1 using NumPy
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
# generate a scatter plot of x and y. The scatter() function takes two arguments: x and y, which are the arrays of random numbers
plt.scatter(x,y)
```

```python
plt.scatter(x,y ,c = "r")
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("This is my x vs y")
```

```python
plt.scatter(x,y ,c = "#790b70") #creates a scatter plot of x and y with purple markers.

#adds a label to the x-axis and y-axis
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("This is my x vs y")
```

```python
plt.scatter(x,y ,c = "#6d1b96" ,alpha =1)
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("This is my x vs y")
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y ,c = "#6d1b96" ,alpha =.9,marker = "*")
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("This is my x vs y")
plt.grid()
```

```python
x= ['a','b','c','d','e']
y= np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.bar(x,y,color = "#1f0b79")
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")

plt.title("This is my bar plot")
plt.grid()
```

```python
plt.figure(figsize=(5,3))
plt.barh(x,y,color = "#814206")
plt.ylabel("representing my cat data")
plt.xlabel("representing my num data")
plt.title("This is my bar plot")
plt.grid()
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8]
plt.figure(figsize=(6,3))
plt.hist(data,color='#1937af')
plt.xlabel("Representing uniqe value")
plt.ylabel("representing feq")
plt.title("Hist plot example")

plt.grid()
plt.savefig("test.png")
```

```python
x = np.random.rand(50)
y= np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z,c= '#1e8106',alpha = .5)
plt.xlabel("my x axis")
plt.ylabel("my x axis")
plt.show()
```
